```python
#Finding the overall average satisfaction score by patient
df[' OverallAverage'] = df[[' StaffSatisfaction', ' CleanlinessSatisfaction', ' FoodSatisfaction', ' ComfortSatisfaction',' CommunicationSatisfaction']].mean(axis=1)

print(df[' OverallAverage'])
```

```python
#importing even more packages
from sklearn.linear_model import LogisticRegression
```

```python
#Logistic Regression 
x = df[' OverallAverage'].values.reshape(-1, 1)
y = df[' Readmission']

log_r = LogisticRegression().fit(x, y)
```

```python
#calculationg correlation
correlation_coefficient = log_r.coef_[0][0]

if correlation_coefficient > 0:
    print("The logistic regression results indicated a: ")
    if correlation_coefficient > 0.5 :
        print("moderate correlation")
    elif correlation_coefficient > 0.7:
        print("strong correlation")
    else:
        print("weak correlation")
else:
    print("The logistic regression results indicated: ")
    print("no orrelation")

print(f"Correlation Coefficient: {correlation_coefficient}.")
```

```python
#plotting logistic regression-fit data
plt.scatter(x, y)
plt.xlabel("Overall Satisfaction")
plt.ylabel("Readmission")
plt.plot(x, log_r.predict(x))
plt.xlim(2, 5)
```
