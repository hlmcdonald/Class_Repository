```python
#Reading the .txt file for Hospital1
print('HOSPITAL 1')
df_1 = pd.read_csv(file_path)
print(df_1)

#Reading the .txt file for Hospital2
print()
print()
print()
print('HOSPITAL 2')
df_2 = pd.read_csv(file_path2)
print(df_2)
```

```python
#Summary Stats for Hosptial 1
print('HOSPITAL 1')
RR = (df_1[' Readmission'].value_counts()[1])
print(f"Number of patients readmitted: {RR}")

AvgStaffSat = np.mean(df_1[' StaffSatisfaction'])
print(f"Average Staff Satisfaction: {AvgStaffSat}")

AvgCleanSat = np.mean(df_1[' CleanlinessSatisfaction'])
print(f"Average Cleanliness Satisfaction: {AvgCleanSat}")

AvgFoodSat = np.mean(df_1[' FoodSatisfaction'])
print(f"Average Food Satisfaction: {AvgFoodSat}")

AvgComfSat = np.mean(df_1[' ComfortSatisfaction'])
print(f"Average Comfort Satisfaction: {AvgComfSat}")

AvgCommSat = np.mean(df_1[' CommunicationSatisfaction'])
print(f"Average Communication Satisfaction: {AvgCommSat}")

#Summary Stats for Hosptial 2
print()
print('HOSPITAL 2')
RR2 = (df_2[' Readmission'].value_counts()[1])
print(f"Number of patients readmitted: {RR2}")

AvgStaffSat2 = np.mean(df_2[' StaffSatisfaction'])
print(f"Average Staff Satisfaction: {AvgStaffSat2}")

AvgCleanSat2 = np.mean(df_2[' CleanlinessSatisfaction'])
print(f"Average Cleanliness Satisfaction: {AvgCleanSat2}")

AvgFoodSat2 = np.mean(df_2[' FoodSatisfaction'])
print(f"Average Food Satisfaction: {AvgFoodSat2}")

AvgComfSat2 = np.mean(df_2[' ComfortSatisfaction'])
print(f"Average Comfort Satisfaction: {AvgComfSat2}")

AvgCommSat2 = np.mean(df_2[' CommunicationSatisfaction'])
print(f"Average Communication Satisfaction: {AvgCommSat2}")
```

```python
#Averaging Hospital 1 Satisfaction Rates

print('HOSPITAL 1')
df_1[' OverallAverage'] = df_1[[' StaffSatisfaction', ' CleanlinessSatisfaction', ' FoodSatisfaction', ' ComfortSatisfaction',' CommunicationSatisfaction']].mean(axis=1)

print(df_1[' OverallAverage'])

print()
print()
print()
print('HOSPITAL 2')
df_2[' OverallAverage'] = df_2[[' StaffSatisfaction', ' CleanlinessSatisfaction', ' FoodSatisfaction', ' ComfortSatisfaction',' CommunicationSatisfaction']].mean(axis=1)

print(df_2[' OverallAverage'])
```

```python
#Importing regression package
from sklearn.linear_model import LogisticRegression

#Logistic Regression for Hospital 1
x = df_1[' OverallAverage'].values.reshape(-1, 1)
y = df_1[' Readmission']

log_reg = LogisticRegression().fit(x, y)

#Correlation Coefficient for Hospital 1
correlation_coefficient = log_reg.coef_[0][0]

if correlation_coefficient > 0:
    print("For hospital 1, the logistic regression results indicated a: ")
    if correlation_coefficient > 0.5 :
        print("moderate correlation")
    elif correlation_coefficient > 0.7:
        print("strong correlation")
    else:
        print("weak correlation")
else:
    print("For hospital 1, the logistic regression results indicated: ")
    print("no correlation")

print(f"Correlation Coefficient: {correlation_coefficient}.")

```

```python
#Data splitting and training
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
#Fitting 
log_r = LogisticRegression()
log_r.fit(x_train, y_train)
```

```python
y_predict = log_r.predict(x_test)
y_predict
```

```python
plt.scatter(x, y)
plt.title('Hospital 1')
plt.xlabel("Overall Satisfaction Scores")
plt.ylabel("Logistic Regression - Overall Satisfaction vs Readmission")
plt.plot(x, log_r.predict(x), color = 'blue')
plt.xlim(1, 5)
```
